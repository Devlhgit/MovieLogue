### 정확성 평가
    
    이 섹션에서는 특정 영화에 대한 추천 결과와 사용자 만족도를 비교하여 정확성을 평가합니다.
    
    1. 사용자가 영화를 선택합니다.
    2. 선택된 영화에 대한 추천 결과를 생성합니다.
    3. 사용자에게 추천된 영화의 만족도를 평가하게 합니다.
    4. 만족도 점수를 바탕으로 정확성을 평가합니다.
    

### 다양성 평가
    
    이 섹션에서는 추천된 영화가 장르, 감독, 배우 등 다양한 측면에서 다양한지 확인합니다.
    
    1. 임의의 영화를 선택하거나 사용자에게 영화를 선택하게 합니다.
    2. 선택된 영화에 대한 추천 결과를 생성합니다.
    3. 추천된 영화의 장르, 감독, 배우 등을 분석하여 다양성을 평가합니다.
